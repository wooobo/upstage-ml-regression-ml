```python
import seaborn as sns
import matplotlib.pyplot as plt

from matplotlib.dates import DateFormatter

import warnings
warnings.filterwarnings('ignore')
```

```python
data = module.file_load.load_current_version()
```

```python
data['is_test'].value_counts()
```

```python
module.parser.날짜포맷_add(data)
module.parser.to_datetime(data, '계약년월일')
module.preprocess.시_군_구_add(data)
```

```python
train = data.query("is_test == 0")
```

```python
df = train.copy()
down_trade_sample1 = df[
    (df["아파트명"] == "염리삼성래미안")
    & (df["계약연도"] >= "2000")
    & (df["전용면적(㎡)"] == 84.96)
]

down_trade_sample2 = df[
    (df["아파트명"] == "헬리오시티")
    & (df["계약연도"] >= "2000")
    & (df["전용면적(㎡)"] >= 84)
    & (df["전용면적(㎡)"] <= 90)
    ]

down_trade_sample3 = df[
    (df["아파트명"] == "현대14차")
    & (df["계약연도"] >= "2000")
    & (df["전용면적(㎡)"] >= 84)
    & (df["전용면적(㎡)"] <= 90)
    ]
```

```python
samples = [down_trade_sample1, down_trade_sample2, down_trade_sample3]

for sample in samples:
    # 라인 차트 생성
    sns.lineplot(x='계약년월일', y='target', data=sample, marker='o', markersize=4)

    plt.title("Down Trade Sample")
    plt.xlabel('Date')
    plt.ylabel('Target (Price)')

    date_formatter = DateFormatter('%Y-%m')
    plt.gca().xaxis.set_major_formatter(date_formatter)

    plt.xticks(rotation=45)

    plt.show()
```
